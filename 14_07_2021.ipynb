```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
what=input("what type of company:-")

url = "https://in.indeed.com/jobs?q="+what
print(url)
html = urlopen(url)
```

```python

soup = BeautifulSoup(html,'lxml')
```

```python
title = soup.title
print(title)
```
